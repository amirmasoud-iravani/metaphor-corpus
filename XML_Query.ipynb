```python
import pandas as pd
from xml.etree.ElementTree import parse, tostring
from google.colab import files

# Parse the XML file
tree = parse('/content/converted_corpus.xml')
root = tree.getroot()

# Ask the user which entries to print
print("Select the Entries You Want to Extract:")
print("1. All Ind_Mets")
print("2. All Imp_Mets")
print("3. All Dir_Mets")
print("4. All M_Flags")
print("5. Ind_Mets with PoS [a= NOUN, PRON, PROPN | b= ADJ, ADV | c= VERB]")
print("6. Imp_Mets with PoS in [a= NOUN, PRON, PROPN | b= ADJ, ADV | c= VERB]")
print("7. Dir_Mets with PoS in [a= NOUN, PRON, PROPN | b= ADJ, ADV | c= VERB]")
print("8. Lex_Unit [a= 'c' | b= 'p' | c= 'n' | d= 'v']")

user_input = input("Enter the Option (1,4,5a,6b ...): ")
selected_entries = []
for x in user_input.split(','):
    if x[-1].isalpha():
        selected_entries.append(int(x[:-1]))
        selected_entries.append(x[-1])
    else:
        selected_entries.append(int(x))

# Print the selected entries to a text file
with open('selected_entries.txt', 'w') as f:
    for entry in root.findall('Entry'):
        should_print = False
        ind_met = entry.find('Ind_Met')
        imp_met = entry.find('Imp_Met')
        dir_met = entry.find('Dir_Met')
        lex_unit = entry.find('Lex_Unit')
        pos = entry.find('PoS')

        if 1 in selected_entries and ind_met is not None and ind_met.text == '2':
            should_print = True
        elif 2 in selected_entries and imp_met is not None and imp_met.text == '2':
            should_print = True
        elif 3 in selected_entries and dir_met is not None and dir_met.text == '2':
            should_print = True
        elif 4 in selected_entries and entry.find('M_Flag') is not None and entry.find('M_Flag').text == '2':
            should_print = True
        elif 5 in selected_entries and ind_met is not None and ind_met.text == '2':
            if selected_entries[selected_entries.index(5) + 1] == 'a' and pos is not None and pos.text in ['NOUN', 'PRON', 'PROPN']:
                should_print = True
            elif selected_entries[selected_entries.index(5) + 1] == 'b' and pos is not None and pos.text in ['ADJ', 'ADV']:
                should_print = True
            elif selected_entries[selected_entries.index(5) + 1] == 'c' and ind_met is not None and ind_met.text == '2' and pos is not None and pos.text == 'VERB':
                should_print = True
        elif 6 in selected_entries and imp_met is not None and imp_met.text == '2':
            if selected_entries[selected_entries.index(6) + 1] == 'a' and pos is not None and pos.text in ['NOUN', 'PRON', 'PROPN']:
                should_print = True
            elif selected_entries[selected_entries.index(6) + 1] == 'b' and pos is not None and pos.text in ['ADJ', 'ADV']:
                should_print = True
            elif selected_entries[selected_entries.index(6) + 1] == 'c' and imp_met is not None and imp_met.text == '2' and pos is not None and pos.text == 'VERB':
                should_print = True
        elif 7 in selected_entries and dir_met is not None and dir_met.text == '2':
            if selected_entries[selected_entries.index(7) + 1] == 'a' and pos is not None and pos.text in ['NOUN', 'PRON', 'PROPN']:
                should_print = True
            elif selected_entries[selected_entries.index(7) + 1] == 'b' and pos is not None and pos.text in ['ADJ', 'ADV']:
                should_print = True
            elif selected_entries[selected_entries.index(7) + 1] == 'c' and dir_met is not None and dir_met.text == '2' and pos is not None and pos.text == 'VERB':
                should_print = True
        elif 8 in selected_entries and lex_unit is not None:
            if selected_entries[selected_entries.index(8) + 1] == 'a' and lex_unit.text == 'c':
                should_print = True
            elif selected_entries[selected_entries.index(8) + 1] == 'b' and lex_unit.text == 'p':
                should_print = True
            elif selected_entries[selected_entries.index(8) + 1] == 'c' and lex_unit.text == 'n':
                should_print = True
            elif selected_entries[selected_entries.index(8) + 1] == 'd' and lex_unit.text == 'v':
                should_print = True

        if should_print:
            f.write(tostring(entry, encoding='unicode'))
            f.write('\n')

print("The selected entries have been saved to 'selected_entries.txt'.")
files.download('selected_entries.txt')
```
