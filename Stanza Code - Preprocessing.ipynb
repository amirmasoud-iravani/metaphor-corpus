```python
# Install the required libraries
!pip install stanza

# Import the necessary libraries
import stanza
from google.colab import files

# Download the Persian model for Stanza
stanza.download('fa')

# Load the Persian model
nlp = stanza.Pipeline('fa')

# Upload the text file
uploaded = files.upload()

# Read the uploaded text file
file_name = next(iter(uploaded))
with open(file_name, 'r', encoding='utf-8') as file:
    text = file.read()

# Perform POS tagging
doc = nlp(text)

# Format the output in CONLL-U format
conll_output = ''
sentence_id = 1
for sent in doc.sentences:
    for word in sent.words:
        conll_output += '\t'.join([str(word.id), word.text, '_', word.upos, word.xpos, '_', str(word.head), word.deprel, '_', '_'])
        conll_output += '\n'
    conll_output += '\n'

# Save the CONLL-U formatted output to a new file
output_file_name = 'output.txt'
with open(output_file_name, 'w', encoding='utf-8') as file:
    file.write(conll_output)

# Download the output file
files.download(output_file_name)
```
